### Part 2: Visualization in 2D and 3D
<b>Step 0: Prepare for plotting</b> <br>
- Import toolkits ในการพล็อตกราฟ 2D และ 3D

```python
import matplotlib.pyplot as plt #2D plotting
from mpl_toolkits.mplot3d import Axes3D #3D plotting
from matplotlib import cm #import color map
```

    ข้อมูลเพิ่มเติม:
    - pyplot: https://matplotlib.org/api/pyplot_api.html
    - mplot3d: https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html

- Import ชุดข้อมูล

```python
import pandas as pd
import numpy as np
df = pd.read_csv('Input.csv') #import ข้อมูล
print(df.head(10)) #แสดงข้อมูลใน df 10 บรรทัดแรก
```

- เก็บข้อมูลแต่ละคอลัมน์ในตัวแปร x1, x2, y

```python
x1 = df.loc[:,'X1']
x2 = df.loc[:,'X2']
y = df.loc[:,'Y']
```

<b>Step 1: Visualize in 2D</b> <br>
- Scatter plot

```python
fig1 = plt.figure() #สร้าง figure ชื่อว่า fig1
ax = fig1.add_subplot(234) #สร้าง subplot
ax.scatter(x1,y)
plt.show()
```

- Line plot: สร้างเส้นกราฟ 2มิติ ตามฟังก์ชันที่เราต้องการ <br>
    สร้าง x-array

```python
xg1 = np.linspace(min(x1),max(x1),200) #สร้าง x-array จำนวน 200 ค่า โดยเริ่มตั้งแต่ค่าต่ำสุดจนถึงสูงสุดของชุดข้อมูล x1 โดยแต่ละค่ามีความห่างเท่าๆกัน
print(xg1)
```

    สร้างค่า y ตามฟังก์ชันที่ต้องการ ยกตัวอย่างจากสมการที่ได้ใน Part1

```python
#Linear regression
y_lin = -118*(xg1)+39000

#Logarithm
y_log = -9990+65800*np.log(xg1)

#Exponential
y_exp = 29300*np.exp(-0.00533*xg1)

#Power
y_pow = 58500*(xg1**(-0.338))
```

    พล็อตกราฟเส้นตามฟังก์ชันที่เรากำหนดไว้

```python
#ยก scatter plot มาจากด้านบน
fig2 = plt.figure()
ax = fig2.add_subplot(111)
ax.scatter(x1,y)

#line plot
ax.plot(xg1,y_lin,color="black") #พล็อตกราฟเส้นสีดำตามฟังก์ชัน y ที่เรากำหนดไว้
#ax.plot(xg1,y_log,color="red")
#ax.plot(xg1,y_exp,color="green")
#ax.plot(xg1,y_pow,color="yellow")
plt.show()
```

    จัดรูปแบบกราฟ

```python
ax.set_xlabel("x1") #ใส่ x-label
ax.set_ylabel("y",rotation=0) #ใส่ y-labelโดยไม่ให้หมุนชื่อแกน
ax.set_title("2D plot of y vs x1")

ax.xaxis.set_label_coords(1.05, -0.025) #กำหนตำแหน่ง x-label ให้ไปอยู่ด้านริม
ax.yaxis.set_label_coords(-0.025, 1.025) #กำหนดตำแหน่ง y-label ให้ไปอยู่บนกราฟ

plt.show() #ความเปลี่ยนแปลงจะแสดงในกราฟด้านบน
```

<b>Step 2: Visualize in 3D</b> <br>
- Scatter plot

```python
fig3 = plt.figure()
ax = fig3.add_subplot(111, projection='3d') #สร้าง subplot ในรูปแบบ 3D
ax.scatter(x1, x2, y, s=10) #กำหนดขนาดจุดเท่ากับ 10
plt.show()
```

- 3D surface plot <br>
สร้าง x,y meshgrid เพื่อกำหนดขอบเขตกราฟ

```python
xg1 = np.linspace(min(x1),max(x1),200) #สร้าง x-array 200 ค่า
xg2 = np.linspace(min(x2),max(x2),200) #สร้าง y-array 200 ค่า
xg1, xg2 = np.meshgrid(xg1, xg2) #นำค่า x, y มาสร้างเป็นตารางของแต่ละ combination ของ x,y
```

    สร้างค่า y โดยการยืมค่าคงที่ที่ได้จาก regression ใน Part1

```python
#Polynomial degree 1
y1 = -115*xg1 + 11500*xg2 +10000
#Polynomial degree 2
y2 = -5*xg1 + 28700*xg2 + 0.385*xg1**2 - 88.2*xg1*xg2 - 858*xg2**2 - 16200
```

    พล็อตลงบน surface

```python
#ยก scatter plot มาจากด้านบน
fig4 = plt.figure()
ax = fig4.add_subplot(111, projection='3d') #สร้าง subplot ในรูปแบบ 3D
ax.scatter(x1, x2, y, s=10) #กำหนดขนาดจุดเท่ากับ 10

#Line plot
#ax.plot_surface(xg1, xg2, y1, cmap=cm.coolwarm, alpha=0.7)
ax.plot_surface(xg1, xg2, y2, cmap = cm.coolwarm, alpha = 0.7) #set color gradient to coolwarm and alpha transparency to 0.7
plt.show()
```

    จัดรูปแบบกราฟ

```python
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("Y")
ax.set_title("Polynomial degree 2: 3D plot of x1, x2 and y")

plt.show() #ความเปลี่ยนแปลงจะแสดงในกราฟด้านบน
```
